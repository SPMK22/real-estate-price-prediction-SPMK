```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the csv file and store it in a variable called data
data = pd.read_csv("immo_data_secondo.csv")

# Set the title and x-axis's title 
plt.title("Corellation between the state of a building and it's price")
plt.xlabel("Price in €")

# Replace all 'None' values with NA, then drop all rows with NA values
data = data.replace('None', pd.NA)
data = data.dropna()

# Group the data by "building_state" and calculate the median price for of them
y = data.groupby("building_state").median()

# Sort the median prices in descending order and save the index as x and the median prices as y
y = y.sort_values(by="Price", ascending=False)["Price"]
x = y.index

# Create a horizontal bar chart with x as the y-axis and y as the x-axis and add some colors
plt.barh(x,y, color ='orange',facecolor='orange')

# For each bar, add the value of the bar above it
for index, value in enumerate(y):
    plt.text(value, index, str("%.0f" % value))

# show the graph 
plt.show()
```
